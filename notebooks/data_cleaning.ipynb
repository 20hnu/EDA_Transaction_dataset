```python
def plot_bar_graph(data,x,y,w,h,tick=False):
  plt.figure(figsize=(w,h))
  ax = sns.barplot(data=data,x=x,y=y,order=data[x])
  for i in ax.containers:
    ax.bar_label(i,)
  plt.ylabel(y+(" (%)"))
  if tick:
    plt.xticks(rotation=45)
  plt.show()
```

```python
def impute_acqCountry(row):
    if pd.isna(row["acqCountry"]):
        if row["merchantCountryCode"] == "US":
            return "US"
        elif row["merchantCountryCode"] == "MEX":
            return "MEX"
        elif row["merchantCountryCode"] == "CAN":
            return "CAN"
        else:
            return "PR"
    return row["acqCountry"]
```

```python
new_df["acqCountry"] = new_df.apply(impute_acqCountry, axis=1)
```

```python
new_df.shape
```

```python
def check(row):
  count = 0
  if row["acqCountry"] != row["merchantCountryCode"]:
    count = 1
  return count
```

```python
new_df['cross_border_count'] = new_df.apply(check, axis=1)
```

```python
# Impute transactionType by highest frequent value
new_df["transactionType"].fillna("PURCHASE", inplace=True)
```

```python
# Rows containing missing values for "posEntryMode" and "posConditionCode" are removed
new_df.dropna(axis=0, inplace=True)
```
